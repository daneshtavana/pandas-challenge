## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)

# Create the age bins in which data will be held
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the four bins
age_bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# add newly formed "Age Group" column to DataFrame using "age_bins" to bin the "Age" column
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=age_bins)

# Same as above, except this time with dupliacte names removed
unique_names_df = purchase_data.drop_duplicates(['SN'])
unique_names_df["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=age_bins)

# group both data frames by newly added "Age Group" column
age_group_df = purchase_data.groupby("Age Ranges")
unique_name_age_group_df = unique_names_df.groupby("Age Ranges")

# Add the calculated columns to the Data frame destined for printing 
df = age_group_df.count()
df['Purchase Count'] = age_group_df.count()[['SN']]
df['Name Count'] = unique_name_age_group_df.count()[['SN']]
df['Total Purchase Value'] = age_group_df.sum()[['Price']]
df['Avg Purchase Price'] = df.apply(lambda x: x['Total Purchase Value'] / x['Purchase Count'], axis=1)
df['Avg Total Purchase per Person'] = df.apply(lambda x: x['Total Purchase Value'] / x['Name Count'], axis=1)

# format and print   
df['Total Purchase Value'] = df['Total Purchase Value'].map("${:,.2f}".format)
df['Avg Purchase Price'] = df['Avg Purchase Price'].map("${:,.2f}".format)
df['Avg Total Purchase per Person'] = df['Avg Total Purchase per Person'].map("${:,.2f}".format)
df[['Purchase Count','Avg Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)

# group data frames by "SN" names column
name_group = purchase_data.groupby("SN")

# Extract and calculated the desired columns to the Data frame 
df = name_group.count()
df['Purchase Count'] = name_group.count()[['Purchase ID']]
df['Total Purchase Value'] = name_group.sum()[['Price']]
df['Avg Purchase Price'] = df.apply(lambda x: x['Total Purchase Value'] / x['Purchase Count'], axis=1)

# Sort by Total Purchase Value   
df = df.sort_values(by='Total Purchase Value', ascending=False)

# format and print  
df['Avg Purchase Price'] = df['Avg Purchase Price'].map("${:,.2f}".format)
df['Total Purchase Value'] = df['Total Purchase Value'].map("${:,.2f}".format)
df[['Purchase Count','Avg Purchase Price','Total Purchase Value']].head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)

# group data frames by "SN" names column
ID_group = purchase_data.groupby(["Item ID", "Item Name"])

# Extract and calculated the desired columns to the Data frame 
df = ID_group.count()
df['Purchase Count'] = ID_group.count()[['Purchase ID']]
df['Total Purchase Value'] = ID_group.sum()[['Price']]
df['Item Price'] = df.apply(lambda x: x['Total Purchase Value'] / x['Purchase Count'], axis=1)

# Sort by Total Purchase Value   
df = df.sort_values(by='Purchase Count', ascending=False)
df_out = df[['Purchase Count','Item Price','Total Purchase Value']]

# format and print  
df['Item Price'] = df['Item Price'].map("${:,.2f}".format)
df['Total Purchase Value'] = df['Total Purchase Value'].map("${:,.2f}".format)
df = df[['Purchase Count','Item Price','Total Purchase Value']]
df.head()
```
